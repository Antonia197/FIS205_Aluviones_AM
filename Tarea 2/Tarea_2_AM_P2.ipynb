#### **Pico de Bragg:** 

![Pico de Bragg](imagenes/Pico%20de%20bragg.jpeg)

Importante mencionar que el gráfico adjunto representa la curva de pico de Bragg para una dosis.

EL pico se debe a como el protón estáinteractuando con el medio, cuando el protón recién ingresa al fantoma, viaja a alta velocidad interactuando rápido y depostiando poca energía por unidad de longitud. Pero a medida que va frenando su velocidad también disminuye drásticamente, provocando que la probabilidad de interactuar e ionizar el medio se dispara justo antes de detenerse por completo. Ahí es donde se genera ese pico característico y localizado de dosis.

#### **Ecuación de Bethe-Bloch:** 

$$ - \frac{dE}{dx}= Kz^2 \frac{Z}{\beta ^2} \left[ \frac{1}{2}ln \frac{2m_e c^2 \beta ^2 \gamma ^2 T_{max}}{I^2} - \beta ^2\right]$$

* **$\frac{dE}{dx}$:** Energía por unidad de longitud
* **K:** Constante de proporcionalidad llamada ''Constante de Bethe-Bloch''.
* **$m_e c^2$:** Constante de energía en reposo del electrón.
* **z:** Carga del proyectil
* **$\beta$:** Velocidad relativa de la partícula 
* **$T_{max}$:** Máxima de transferencia de energía
* **$Z$:** Número atómico promedio del medio.
* **$A$:** Masa atómica promedio del medio.
* **$I$:** Potencial medio de exitación del medio
* **$\gamma$:** Factor de Lorentz


##### **Origen de $z^2$ y $\frac{1}{\beta^2}$**

* **$z^2$ (La carga):** Viene de la fuerza de Coulomb. Al duplicar la carga de la partícula, la fuerza sobre el electrón se duplica. Como la energía transferida depende del cuadrado del impulso ($\Delta E \propto \Delta p^2$), la pérdida de energía termina dependiendo de $z^2$.

* **$1/\beta^2$ (La velocidad):** Viene del tiempo de interacción. Una partícula que viaja muy rápido pasa menos tiempo cerca del electrón, por lo que el "empujón" eléctrico dura menos y le transfiere menos energía. Al frenarse (menor $\beta$), el tiempo de interacción aumenta exponencialmente, haciendo que la pérdida de energía se dispare justo al final de su trayecto.

#### **Energy straggling:**

La ecuación de Bethe-Bloch te da un promedio determinista (como si todos los protones se frenaran exactamente igual). Pero el proceso real es estocástico (probabilístico): un protón puede chocar con más electrones que otro por pura suerte, a esto se le llama **Energy Straggling**.

#### **Dispersión Coulombiana Múltiple:**

En los términos anteriores solo consideramos que el protón avanza en línea recta perdiendo energía con los electrones. Pero el medio posee núcleos atómicos cargados positivamente. Esto quiere decir que cuando el protón pasa cerca de un núcleo, se desvía un ángulo pequeño debido al rechazo coulombiano. Al ocurrir miles de veces, el protón no viaja en línea recta, si no una trayectoria zigzagueante.

* **Ensanchamiento del pico y caída de dosis:** En el caso ideal (siendo el inciso c), todos los protones viajan acoplados y se detienen en el mismo punto exacto, generando la cresta de dosis infinita. Pero al incluir la fluctuación probabilística de Bohr (inciso d), cada protón experimenta colisiones distintas. Provocando que el pico de Bragg se ensanche y baje su altura, ya que la energía se reparte.

* **Importancia clínica:** El straggling demuestra que en física médica un haz monoenergético no es perfectamente nítido. Para tratar tumores reales de forma homogénea, es obligatorio superponer haces de distintas energías para diseñar un pico plano que cubra toda la lesión (Spread-Out Bragg Peak o SOBP).

```python
import numpy as np
from scipy.integrate import quad

# ==========================================
# 1. CONSTANTES (Unidades: MeV, cm)
# ==========================================
me_c2 = 0.511        # Masa del electrón (MeV)
Mp_c2 = 938.27       # Masa del protón (MeV)
z = 1                # Carga del protón
I = 75e-6            # Potencial de excitación del agua (MeV)
rho = 1.0            # Densidad del agua (g/cm^3)
Z_over_A = 10 / 18   # Relación Z/A para el agua (H2O)
K = 0.307075         # Constante de Bethe-Bloch (MeV * cm^2 / mol)

# ==========================================
# 2. PODER DE FRENADO (-dE/dx)
# ==========================================
def dedx_bethe_bloch(E):
    """
    Calcula la pérdida de energía por unidad de longitud (-dE/dx) en MeV/cm
    para un protón con energía cinética E (MeV) en agua.
    """
    # Evitar divisiones por cero o energías negativas
    if E <= 0:
        return float('inf')
    
    # Cinemática relativista
    gamma = 1.0 + (E / Mp_c2)
    beta_quad = 1.0 - (1.0 / (gamma**2))
    
    # Máxima transferencia de energía aproximada
    T_max = 2 * me_c2 * beta_quad * (gamma**2)
    
    # Términos dentro del corchete de Bethe-Bloch
    log_term = 0.5 * np.log((2 * me_c2 * beta_quad * T_max) / (I**2))
    corchete = log_term - beta_quad
    
    # Ecuación completa de Bethe-Bloch (-dE/dx) multiplicada por la densidad
    pérdida = rho * K * (z**2) * Z_over_A * (1.0 / beta_quad) * corchete
    return pérdida

# ==========================================
# 3. INTEGRACIÓN PARA RANGO CSDA Y COMIENZO DEL SCRIPT
# ==========================================
if __name__ == "__main__":
    # Solicitar al usuario la energía cinética inicial E_0
    try:
        E0_input = float(input("Ingrese la energía inicial del protón E_0 en MeV (ej. 50, 150, 250): "))
        
        # El integrando es 1 / (-dE/dx). Usamos una función lambda para llamarlo.
        # Integramos desde un valor muy cercano a cero (ej. 1e-3) hasta E_0 para evitar la singularidad en E=0.
        rango_cm, error = quad(lambda E: 1.0 / dedx_bethe_bloch(E), 0.05, E0_input, limit=100)
        
        print("\n--- RESULTADOS ---")
        print(f"Para E_0 = {E0_input} MeV:")
        print(f"Rango CSDA calculado: {rango_cm:.4f} cm ({rango_cm * 10:.2f} mm)")
        
    except ValueError:
        print("Por favor, ingrese un número válido para la energía.")

#Datos de PSTAR del NIST para comparación:
# E_0 (MeV) | Rango CSDA (cm)
# 50        | 2.227
# 150       | 15.770
# 250       | 37.940




import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Aseguramos que la función de rango por integración esté definida en esta celda
def calcular_rango_csda(E0):
    """Calcula el rango CSDA integrando numéricamente 1/(-dE/dx)."""
    rango_cm, _ = quad(lambda E: 1.0 / dedx_bethe_bloch(E), 0.05, E0, limit=100)
    return rango_cm

def simular_trayectoria_csda(E0, dx=0.01):
    """
    Simula el frenado continuo de un protón desde z=0 hasta detenerse.
    Retorna los arreglos de posiciones z (cm) y la energía depositada (MeV) en cada paso.
    """
    posiciones_z = []
    energia_depositada = []
    
    z_actual = 0.0
    E_actual = E0
    
    # El protón avanza mientras le que de energía cinética
    while E_actual > 0:
        dedx = dedx_bethe_bloch(E_actual)
        dE = dedx * dx
        
        # Control de frontera: no puede perder más energía de la que tiene
        if dE >= E_actual:
            dE = E_actual  # Deposita el remanente exacto
            E_actual = 0.0
        else:
            E_actual -= dE
            
        posiciones_z.append(z_actual)
        energia_depositada.append(dE)
        z_actual += dx
        
    return np.array(posiciones_z), np.array(energia_depositada)

# Executar la simulación para el Inciso C
E0 = 150.0          # Energía inicial (MeV) [cite: 66]
N_protones = 10000  # Número de protones [cite: 66]
dx = 0.01           # Paso espacial de 0.1 mm expresado en cm [cite: 66]

# Ejecutar el transporte del protón representativo
z_pasos, dE_per_proton = simular_trayectoria_csda(E0, dx)

# Multiplicar por N para obtener la energía total depositada por el haz completo
dosis_total = dE_per_proton * N_protones

# Convertir a unidades de Dosis Relativa por unidad de longitud (MeV/cm)
dosis_por_cm = dosis_total / dx

# ==========================================
# GRAFICAR LA DOSIS D(z) Y EL PICO DE BRAGG
# ==========================================
plt.figure(figsize=(9, 5))

# Graficar la dosis calculada paso a paso
plt.plot(z_pasos, dosis_por_cm, color='darkblue', lw=2.5, label='Dosis $D(z)$ (Simulación paso a paso)')

# Calcular el rango analítico del inciso b para comparar
rango_b = calcular_rango_csda(E0)
plt.axvline(x=rango_b, color='crimson', linestyle='--', lw=1.5, 
            label=f'Rango $R_{{CSDA}}$ Inciso b ({rango_b:.2f} cm)')

# Ajustes de formato y etiquetas del gráfico
plt.title('Dosis Depositada vs Profundidad $z$ (Pico de Bragg Teórico)', fontsize=12, fontweight='bold')
plt.xlabel('Profundidad $z$ (cm)', fontsize=11)
plt.ylabel('Dosis Deposita / Unidad de longitud (MeV/cm)', fontsize=11)
plt.xlim(0, rango_b + 1.0) 
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(loc='upper left')

plt.show()


# ==============================================================================
# INCISO D: SIMULACIÓN MONTE CARLO CON ENERGY STRAGGLING
# ==============================================================================



def simular_monte_carlo_straggling(E0, N_protons, dx=0.01):
    """
    Simula el transporte estocástico de N protones incorporando la fluctuación
    de Bohr (Energy Straggling). Acumula la energía en una rejilla espacial fija.
    """
    # Definimos una rejilla espacial fija desde 0 hasta 18 cm para acumular la dosis
    z_max_grid = 18.0
    bins_z = np.arange(0, z_max_grid + dx, dx)
    dosis_acumulada = np.zeros(len(bins_z) - 1)
    
    print(f"Iniciando simulación Monte Carlo para {N_protons} protones...")
    
    for i in range(N_protons):
        E_actual = E0
        z_actual = 0.0
        
        # Monitorear el progreso cada 2000 protones para no desesperar
        if i % 2000 == 0 and i > 0:
            print(f" -> Protones simulados: {i}/{N_protons}")
            
        while E_actual > 0:
            # 1. Obtener el frenado medio
            dedx = dedx_bethe_bloch(E_actual)
            mean_dE = dedx * dx
            
            # 2. Calcular la varianza de Bohr (sigma^2) para este paso
            gamma = 1.0 + (E_actual / Mp_c2)
            beta_quad = 1.0 - (1.0 / (gamma**2))
            
            # Expresión simplificada de sigma_E^2 acoplada a nuestras constantes
            sigma_E_quad = rho * K * (z**2) * Z_over_A * (1.0 / beta_quad) * dx * (2 * me_c2 * beta_quad * gamma**2)
            sigma_E = np.sqrt(max(0.0, sigma_E_quad))
            
            # 3. Muestrear la pérdida de energía real usando una Normal
            if sigma_E > 0:
                dE_real = np.random.normal(mean_dE, sigma_E)
            else:
                dE_real = mean_dE
                
            # 4. Control de fronteras físicas
            if dE_real >= E_actual or E_actual < 0.1: 
                dE_real = E_actual
                E_actual = 0.0
            else:
                E_actual -= dE_real
            
            # 5. Identificar en qué celda espacial se encuentra el protón y sumar la energía
            idx = int(z_actual / dx)
            if idx < len(dosis_acumulada):
                dosis_acumulada[idx] += dE_real
                
            z_actual += dx
            
    # Retornamos el centro de cada celda (z) y la dosis total normalizada por unidad de longitud
    z_centros = bins_z[:-1] + dx/2
    dosis_por_cm = dosis_acumulada / dx
    
    return z_centros, dosis_por_cm

# Ejecutar la simulación probabilística si corres el script
if __name__ == "__main__":
    # El código que ya tenías de los incisos b y c corre primero...
    
    # Lanzamos el cálculo estocástico (puede tardar unos segundos)
    z_mc, dosis_mc = simular_monte_carlo_straggling(E0=150.0, N_protons=10000, dx=0.01)
    
    # ==============================================================================
    # GRÁFICO COMPARATIVO FINAL: INCISO C VS INCISO D
    # ==============================================================================
    plt.figure(figsize=(10, 6))
    
    # Curva del inciso C (Sin straggling - la aguja que ya obtuviste)
    plt.plot(z_pasos, dosis_por_cm, color='darkblue', lw=2, label='Sin Straggling (CSDA Puro)')
    
    # Curva del inciso D (Con straggling - Monte Carlo)
    plt.plot(z_mc, dosis_mc, color='crimson', lw=2, label='Con Straggling (Monte Carlo - Bohr)')
    
    plt.title('Efecto del Energy Straggling en el Pico de Bragg (E0 = 150 MeV)', fontsize=12, fontweight='bold')
    plt.xlabel('Profundidad z (cm)', fontsize=11)
    plt.ylabel('Dosis Depositada (MeV/cm)', fontsize=11)
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend(loc='upper left')
    
    plt.show()
```
